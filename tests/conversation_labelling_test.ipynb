```python
import os
import json

base_dir = '../example_conversations'
language = 'english'

with open(os.path.join(base_dir, language, 'plain_text_conv_1.txt'), 'r') as f:
    plain_text = f.read()

with open(os.path.join(base_dir, language, 'labelled_text_conv_1.json'), 'r') as f:
    json_text = json.loads(f.read())

plain_text, json_text
```

```python
line_splits = plain_text.split('\n')

plain_tokens = []

punctuation = '.,?!:'

for line in line_splits:
    for punct in punctuation:
        line = line.replace(punct, '')
    plain_tokens.extend(line.split(' '))

plain_tokens
```

```python
for i, jtok in enumerate(json_text):
    if list(jtok.keys())[0] != plain_tokens[i]:
        print(f'ERROR at {i}: {jtok} ; {plain_tokens[i]}')
        break
```

```python
from embeddings import get_embeddings
import numpy as np

single_word_embeddings = np.array(get_embeddings(plain_tokens))

np.save(os.path.join(base_dir, language, 'conv_1_word.npy'), single_word_embeddings)
```

```python
roll_tokens = []
roll = 3
for i in range(len(plain_tokens)):
    if i + roll >= len(plain_tokens):
        idx = len(plain_tokens)
    else:
        idx = i + roll
    roll_tokens.append(' '.join(plain_tokens[i:idx]))

windowed_embeddings_roll = np.array(get_embeddings(roll_tokens))

np.save(os.path.join(base_dir, language, f'conv_1_roll{roll}.npy'), windowed_embeddings_roll)
```
